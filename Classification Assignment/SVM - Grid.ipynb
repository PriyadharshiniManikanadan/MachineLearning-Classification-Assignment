## SVM - Grid

```python
import pandas as pd
```

### Data Collection

```python
dataset = pd.read_csv("CKD.csv")
dataset
```

### Data Preprocessing

```python
dataset = pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
Class_count = dataset["classification_yes"].value_counts()
Class_count
```

### Input Output Split

```python
indep = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv','wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal', 'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes', 'appet_yes', 'pe_yes', 'ane_yes']]
dep = dataset[['classification_yes']]
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(indep,dep,test_size=0.2,random_state=0)
```

### Standardization

```python
from sklearn.preprocessing import StandardScaler
SC = StandardScaler()
X_train = SC.fit_transform(X_train)
X_test = SC.transform(X_test)
```

### Model Creation 

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

param_grid = {'kernel':['linear','rbf','poly','sigmoid'],
              'C':[1,10,100,1000,2000,3000],
              'gamma':['scale','auto']}

Classifier = GridSearchCV(SVC(probability = True), param_grid, refit=True, verbose=3, n_jobs=-1, scoring='f1_weighted')
Classifier.fit(X_train,Y_train)
```

### Test Set

```python
grid_results = Classifier.cv_results_

grid_predictions = Classifier.predict(X_test)
```

### Evaluation Metrics

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
CM = confusion_matrix(Y_test,grid_predictions)
CM
```

### Classification Report

```python
from sklearn.metrics import classification_report
Clf_report = classification_report(Y_test,grid_predictions)
print(Clf_report)
```

### ROC_AUC_SCORE

```python
from sklearn.metrics import roc_auc_score
ROC_Score = roc_auc_score(Y_test,Classifier.predict_proba(X_test)[:,1])
ROC_Score
```

### F1_SCORE

```python
from sklearn.metrics import f1_score
f1_macro = f1_score(Y_test,grid_predictions,average='macro')
print("Best Params:", format(Classifier.best_params_),f1_macro)
```
